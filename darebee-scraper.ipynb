```python
# imports necessary python modules

import webbrowser
from bs4 import BeautifulSoup
import requests
from urllib.request import Request, urlopen
import pandas as pd
from numpy import nan
```

```python
timeList = []
textList = []
avatarList = []
titleList = []
fromList = []
postsList = []
```

```python
# link = "https://darebee.com/hive/logs/243334-a-wolf-s-steps-through-the-snow"
link = "https://darebee.com/hive/logs/243334-a-wolf-s-steps-through-the-snow/page7"
headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}

page = requests.get(link, headers=headers)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
dElems = soup.select('div')
tElems = soup.select('time')
```

```python
# appends datetimes
for elem in tElems:
    timeList.append(elem.attrs['datetime'])
```

```python
# appends text
for elem in dElems:
    if 'class' in elem.attrs:
        if elem.attrs['class'] == ['js-post__content-text', 'OLD__post-content-text', 'restore', 'h-wordwrap']:
            textList.append(elem.getText().strip().encode('utf-8'))
```

```python
# appends bio
bioList = []
for elem in dElems:
    if 'class' in elem.attrs:
        if elem.attrs['class'] == ['dareProfile_left']:
            bioList.append(elem.getText().strip())
```

```python
# to strip the avatar details 
stripList = []
for row in bioList:
    text = row.replace(' ', '').replace("\r", ' ').replace('\n', ' ')
    text = text.split()
    stripList.append(text)
```

```python
# puts appropriate NaN values
for l in stripList:
    avatarList.append(l[0])
    postsList.append(l[len(l) - 1][6:])
    if len(l) == 4:
        titleList.append(l[1])
        fromList.append(l[2][4:])      
    elif len(l) == 3:
        if l[1].startswith("from"):
            fromList.append(l[1][4:])
            titleList.append(None)
        else:
            titleList.append(l[1])
            fromList.append(None)
    else:
        titleList.append(None)
        fromList.append(None)
```

```python
bigDict = {'avatar': avatarList, 'title': titleList, 'from': fromList, 'posts': postsList, 'time': timeList, 'text': textList}
```

```python
df = pd.DataFrame(bigDict)
df.fillna(value=nan, inplace=True)
```
